```python
import pandas as pd
import numpy as np
```

## 1. Data Loading

**Membaca Dataset**
- `pd.read_excel()`: fungsi untuk membaca file Excel (.xls atau .xlsx)
- Data dari file 'CADalizadeh.xls' dimuat ke dalam DataFrame dengan nama variabel `df`
- DataFrame adalah struktur data 2 dimensi (tabel) yang digunakan pandas untuk menyimpan data

```python
df = pd.read_excel('CADalizadeh.xls')
```

## 2. Data Exploration

**1. Menampilkan 5 Baris Pertama (`df.head()`)**
- Menampilkan 5 baris awal dari dataset
- Berguna untuk melihat struktur data dan contoh isi dataset

**2. Informasi Dataset (`df.info()`)**
- Menampilkan jumlah kolom dan baris
- Tipe data setiap kolom (int, float, object, dll)
- Jumlah nilai non-null (tidak kosong) di setiap kolom
- Penggunaan memori dataset

**3. Statistik Deskriptif (`df.describe()`)**
- Menampilkan statistik untuk kolom numerik:
  - **count**: jumlah data
  - **mean**: nilai rata-rata
  - **std**: standar deviasi (sebaran data)
  - **min**: nilai minimum
  - **25%, 50%, 75%**: kuartil (percentil)

**4. Shape dan Nama Kolom**
- **Shape**: dimensi dataset (jumlah baris × jumlah kolom)

Tujuan dari data exploration ini adalah untuk memahami karakteristik dataset sebelum melakukan data cleaning dan preprocessing lebih lanjut.

```python
print("Data Head:")
print(df.head())
print("\n" + "="*50 + "\n")

print("Informasi Dataset:")
print(df.info())
print("\n" + "="*50 + "\n")

print("Statistik Deskriptif:")
print(df.describe())
print("\n" + "="*50 + "\n")

print(f"Shape Dataset: {df.shape}")
print(f"Jumlah Baris: {df.shape[0]}")
print(f"Jumlah Kolom: {df.shape[1]}")
print("\nNama Kolom:")
print(df.columns.tolist())

print("Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0])

print("\n" + "="*50 + "\n")
print("Persentase Missing Values:")
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)
if len(missing_percent) > 0:
    print(missing_percent)
else:
    print("Tidak ada missing values dalam dataset")

print("\n" + "="*50 + "\n")
print("Correlation Matrix:")
numeric_cols = df.select_dtypes(include=[np.number]).columns
if len(numeric_cols) > 1:
    correlation = df[numeric_cols].corr()
    print(correlation)
else:
    print("Tidak cukup kolom numerik untuk membuat correlation matrix")

print("\n" + "="*50 + "\n")
print("Jumlah Kolom Numerik:", len(df.select_dtypes(include=[np.number]).columns))
print("Kolom Numerik:", df.select_dtypes(include=[np.number]).columns.tolist())

print("\n" + "="*50 + "\n")
print("Jumlah Kolom Kategorikal:", len(df.select_dtypes(include=['object']).columns))
print("Kolom Kategorikal:", df.select_dtypes(include=['object']).columns.tolist())

categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    for col in categorical_cols[:5]:
        value_counts = df[col].value_counts()
        print(f"\nDistribusi '{col}':")
        print(value_counts)
else:
    print("Tidak ada kolom kategorikal dalam dataset")
```

## 3. Data Cleaning

**Tujuan Preprocessing Data**
- Memeriksa dan menangani missing values (data kosong)
- Memastikan kualitas data sebelum proses lebih lanjut

**Metode Handling Missing Values:**
- **Opsi 1 - Drop**: Menghapus baris yang memiliki missing values
- **Opsi 2 - Imputation**: Mengisi missing values dengan nilai tertentu
  - Kolom numerik: diisi dengan **mean** (rata-rata)
  - Kolom kategorikal: diisi dengan **mode** (nilai terbanyak)

**Output:**
- `df_cleaned`: DataFrame setelah dibersihkan
- Menampilkan jumlah missing values sebelum dan sesudah cleaning

```python
print("Jumlah missing values sebelum cleaning:")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")

df_cleaned = df.copy()
for col in df_cleaned.columns:
    if df_cleaned[col].isnull().sum() > 0:
        if df_cleaned[col].dtype in ['float64', 'int64']:
            df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
        else:
            df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

print("Jumlah missing values setelah cleaning:")
print(df_cleaned.isnull().sum())
print(f"\nShape setelah cleaning: {df_cleaned.shape}")
```

## 4. Data Encoding

**Tujuan:**
- Mengubah semua kolom menjadi bernilai numerik
- Algoritma machine learning hanya dapat memproses data numerik

**Metode Label Encoding:**
- Mengubah kategori menjadi angka (0, 1, 2, ...)
- Menggunakan `LabelEncoder` dari sklearn
- Cocok untuk variabel kategorikal ordinal

**Proses:**
- Identifikasi kolom bertipe object (kategorikal)
- Transform setiap kolom kategorikal menjadi numerik
- Simpan encoder untuk kemungkinan inverse transform

**Output:**
- `df_encoded`: DataFrame dengan semua kolom numerik
- `label_encoders`: Dictionary berisi encoder untuk setiap kolom

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
print(f"Kolom kategorikal yang akan di-encode: {list(categorical_columns)}")
print("\n" + "="*50 + "\n")

df_encoded = df_cleaned.copy()

label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le
    print(f"Kolom '{col}' berhasil di-encode")
    print(f"  - Kategori unik: {len(le.classes_)}")
    print(f"  - Classes: {le.classes_[:5]}..." if len(le.classes_) > 5 else f"  - Classes: {le.classes_}")
    print()

print("="*50)
print("\nTipe data setelah encoding:")
print(df_encoded.dtypes)
print(f"\nSemua kolom sudah numerik: {df_encoded.select_dtypes(include=['object']).shape[1] == 0}")
```

## 5. Data Normalization

**Tujuan:**
- Mengubah skala nilai atribut ke range yang sama
- Mencegah atribut dengan nilai besar mendominasi model
- Mempercepat konvergensi algoritma machine learning

**Metode Min-Max Normalization:**
- Mengubah nilai ke range [0, 1]
- Formula: `(x - min) / (max - min)`
- Cocok untuk data dengan distribusi tidak normal

**Metode Standardization (Z-score):**
- Mengubah nilai dengan mean=0 dan std=1
- Formula: `(x - mean) / std`
- Cocok untuk algoritma yang mengasumsikan distribusi normal

**Output:**
- `df_normalized_minmax`: DataFrame dengan Min-Max Normalization
- `df_normalized_standard`: DataFrame dengan Standardization
- `df_final`: Dataset final yang siap digunakan untuk modeling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

print("Statistik sebelum normalisasi:")
print(df_encoded.describe())
print("\n" + "="*50 + "\n")

scaler_minmax = MinMaxScaler()
df_normalized_minmax = pd.DataFrame(
    scaler_minmax.fit_transform(df_encoded),
    columns=df_encoded.columns,
    index=df_encoded.index
)

print("Hasil Min-Max Normalization (range 0-1):")
print(df_normalized_minmax.describe())
print("\n" + "="*50 + "\n")

scaler_standard = StandardScaler()
df_normalized_standard = pd.DataFrame(
    scaler_standard.fit_transform(df_encoded),
    columns=df_encoded.columns,
    index=df_encoded.index
)

print("Hasil Standardization (mean=0, std=1):")
print(df_normalized_standard.describe())
print("\n" + "="*50 + "\n")

df_final = df_normalized_minmax.copy()
print(f"Dataset final shape: {df_final.shape}")
print(f"Dataset final (5 baris pertama):")
print(df_final.head())
```
